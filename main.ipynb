# Step 2: VADER Sentiment Scoring Model

```python
from nltk.sentiment import SentimentIntensityAnalyzer
from tqdm.notebook import tqdm

sa = SentimentIntensityAnalyzer()
```

```python
sa.polarity_scores('I am sad.')
```

```python
sa.polarity_scores('I am happy, but this is worst than ever.')
```

```python
sa.polarity_scores(example)
```

```python
#Running the polarity score on the entire dataset
result = {}
for i,row in tqdm(df.iterrows(),total=len(df)):
    text = row['Text']
    myid = row['Id']
    result[myid] = sa.polarity_scores(text)
```

```python
vaders = pd.DataFrame(result).T
vaders = vaders.reset_index().rename(columns={'index':'Id'})
vaders = vaders.merge(df, how='left')
```

```python
vaders.head()
```

```python
ax = sns.barplot(data = vaders, x = 'Score', y= 'compound')
ax.set_title('Compound Score by Amazon Star Review')
plt.show()
```

```python
fig, ax1 = plt.subplots(1,3,figsize=(12,5))
sns.lineplot(data = vaders, x='Score',y='pos',ax=ax1[0])
sns.lineplot(data = vaders, x='Score',y='neu',ax=ax1[1])
sns.lineplot(data = vaders, x='Score',y='neg',ax=ax1[2])
ax1[0].set_title('Positive')
ax1[1].set_title('Neutral')
ax1[2].set_title('Negative')
plt.tight_layout()
plt.show()
```

```python
'''VADER Model looks for each word in the sentence and scores each word 
individually with a sentiment analysis score based on the models complexity.
But the human language is based on context and scoring each words individually
can sometime result in a mismatch in the sentiment of the entire sentence. 
So we opt for a more powerful model of BERT to get a better context of the
entire sentence structure and the emotion in the text.''' 
```

# Step 3: RoBERTa Pre-trained Model

```python
!pip install transformers
```

```python
from transformers import AutoTokenizer
from transformers import AutoModelForSequenceClassification
from scipy.special import softmax
```

```python
MODEL = f"cardiffnlp/twitter-roberta-base-sentiment"
tokenizer = AutoTokenizer.from_pretrained(MODEL)
model = AutoModelForSequenceClassification.from_pretrained(MODEL)
```

```python
print(example)
sa.polarity_scores(example)
```

```python
#Running the RoBERTa Model
encoded_text = tokenizer(example,return_tensors='pt')
output = model(**encoded_text)
scores = output[0][0].detach().numpy()
scores = softmax(scores)
scores_dict = {
    'roberta_neg' : scores[0],
    'roberta_neu' : scores[1],
    'roberta_pos' : scores[2]
}
print(scores_dict)
```

```python
def polarity_scores_roberta(example):
    encoded_text = tokenizer(example,return_tensors='pt')
    output = model(**encoded_text)
    scores = output[0][0].detach().numpy()
    scores = softmax(scores)
    scores_dict = {
        'roberta_neg' : scores[0],
        'roberta_neu' : scores[1],
        'roberta_pos' : scores[2]
}
    return scores_dict
    
```

```python
result = {}
for i,row in tqdm(df.iterrows(),total=len(df)):
    try:
        text = row['Text']
        myid = row['Id']
        vader_result = sa.polarity_scores(text)
        vader_result_rename = {}
        for key,value in vader_result.items():
            vader_result_rename[f"vader_{key}"] = value
        roberta_result = polarity_scores_roberta(text)
        comb = {**vader_result_rename, **roberta_result}
        result[myid]=comb
    except RuntimeError:
        print(f'Broke for id {myid}')
    
    
```

```python
result_df = pd.DataFrame(result).T
result_df = result_df.reset_index().rename(columns={'index':'Id'})
result_df = result_df.merge(df,how='left')
```
